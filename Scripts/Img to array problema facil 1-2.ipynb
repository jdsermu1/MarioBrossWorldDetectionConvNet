```
import numpy as np
import os
import cv2
```

```
PATH = os.getcwd()
data_path = PATH + '/drive/My Drive/Colab Notebooks/Machine Learning/Proyecto 1/Imagenes'
data_dir_list = os.listdir(data_path)
print(data_dir_list)
img_data_list=[]
```

```
y=[]
n=0
for dataset in data_dir_list:
  stages_list=os.listdir(data_path+'/'+ dataset)
  print ('Loading the images of dataset-'+'{}\n'.format(dataset))
  if dataset=='Mundo 1':
    for stage in stages_list:
      img_list=os.listdir(data_path+'/'+ dataset+'/'+stage)
      n=0
      for img in img_list:
        if n<=1034:
          input_img=cv2.imread(data_path + '/'+ dataset+'/'+stage + '/'+ img)
          input_img_resize=cv2.resize(input_img,(256,256))
          #cv2_imshow(input_img_resize)
          img_data_list.append(input_img_resize)
          y.append(1)
        n+=1
  if dataset=='Mundo 2':
    for stage in stages_list:
      img_list=os.listdir(data_path+'/'+ dataset+'/'+ stage)
      n=0
      for img in img_list:
        if n<=1034:
          input_img=cv2.imread(data_path + '/'+ dataset+'/'+stage + '/'+ img)
          input_img_resize=cv2.resize(input_img,(256,256))
          #cv2_imshow(input_img_resize)
          img_data_list.append(input_img_resize)
          y.append(0)
        n+=1
```

```
datos=np.array(img_data_list)
print(datos.shape)
y=np.array(y)
print(y.shape)
cad=PATH + '/drive/My Drive/Colab Notebooks/Machine Learning/Proyecto 1/Arreglos/Problema Facil/'
np.save(cad+"Xnumpyarray", datos)
np.save(cad+"Ynumpyarray", y)
```

```
zeros=0
ones=0
for i in y:
  if i==0:
    zeros+=1
  if i==1:
    ones+=1
print('zeros: ',zeros)
print('ones: ',ones)
```
